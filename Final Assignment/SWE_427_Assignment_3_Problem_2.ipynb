```python
# Dropping this column because bank statement is less likely to be effected by these incidents
df = pd.read_csv('bank-full.csv', delimiter=';').drop([
    'default', 'contact', 'day', 'month', 'pdays', 'previous', 'poutcome'
], axis=1)
df
```

```python
print(
  # df['age'].unique(),
  df['job'].unique(),
  df['marital'].unique(),
  df['education'].unique(),
  df['housing'].unique(),
  df['y'].unique(),
  # df['campaign'].unique()
)
```

```python
le = LabelEncoder()

df['job'] = le.fit_transform(df['job'])
df['marital'] = le.fit_transform(df['marital'])
df['education'] = le.fit_transform(df['education'])
df['housing'] = le.fit_transform(df['housing'])
df['y'] = le.fit_transform(df['y'])
df['loan'] = le.fit_transform(df['loan'])
```

```python
print(
  # df['age'].unique(),
  df['job'].unique(),
  df['marital'].unique(),
  df['education'].unique(),
  df['housing'].unique(),
  df['y'].unique(),
  # df['campaign'].unique()
)
```

```python
df
```

```python
Y = df['y']
X = df.drop('y', axis=1)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

model = SVC(kernel='rbf', gamma='auto')
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)

accuracy = accuracy_score(Y_test, Y_pred)
print('Model Accuracy is:', str(math.floor(accuracy * 100)) + '%')
```
